## **Question 1: Calculate Total Annual Revenue**

**Revenue = Units Produced × Price per Unit**

```python
revenue = units_per_year * price_per_unit

print("Total Annual Revenue:", revenue)
```

## **Question 2: Net Benefit**

**Net Benefit = Total Revenue – Annual Operational Cost – Initial Investment**

```python
net_benefit = revenue - operational_cost - initial_investment

print("Net Benefit:", net_benefit)
```

## **Question 4: Generating Synthetic Production Data**

**Since real manufacturing data is not available, synthetic data can be generated using Python.**

**This simulated dataset represents realistic manufacturing conditions.**

```python
np.random.seed(42)

n = 500

production_time = np.random.normal(5,0.8,n)

material_strength = np.random.normal(80,4,n)

weight_reduction = 0.35 * material_strength + np.random.normal(0,2,n)

defect_rate = 1 + (production_time * 0.5) + np.random.normal(0,0.5,n)

production_cost = 3500 + (production_time * 300) + (defect_rate * 200)

data = pd.DataFrame({

"Production_Time": production_time,
"Material_Strength": material_strength,
"Weight_Reduction_%": weight_reduction,
"Defect_Rate_%": defect_rate,
"Production_Cost": production_cost

})

data.head()
```

**Production Cost Distribution**

```python
plt.figure(figsize=(8,5))

sns.histplot(data['Production_Cost'], bins=30)

plt.title("Distribution of Production Cost")

plt.show()
```

**This graph shows how production costs are distributed across different manufacturing batches.**

**Production Time vs Cost**

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
x=data['Production_Time'],
y=data['Production_Cost']
)

plt.title("Production Time vs Production Cost")

plt.show()
```

**Longer production times generally increase manufacturing costs.**

**Defect Rate Analysis**

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=data['Defect_Rate_%'])

plt.title("Defect Rate Distribution")

plt.show()
```
